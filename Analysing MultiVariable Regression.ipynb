In this project I have loaded boston dataset split it into train and test and applied linear regression on it. Then played with
the features i.e taking square of features as an extra column or taking product of any two feature as an extra column and then 
analysed the effect of it on my score 

```python
from sklearn import datasets
import pandas as pd
```

```python
boston = datasets.load_boston()
```

```python
X = boston.data
Y = boston.target
print(X.shape,Y.shape)
```

```python
df = pd.DataFrame(X)
df.columns = boston.feature_names
df1 = df.copy()
```

```python
for i in boston.feature_names:
    df[i*2] = (df[i])**2
```

df1 is the pandas dataframe of features of boston data whereas the df dataframe created in the above cell contain all the features of boston data and it contains the square of all features as extra feature.

```python
df1
```

```python
df
```

```python
X1 = df.values
X1.shape
```

```python
from sklearn import model_selection
```

```python
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,random_state = 0)
X1_train,X1_test,Y1_train,Y1_test = model_selection.train_test_split(X1,Y,random_state = 0)
X.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
alg1 = LinearRegression()
alg1.fit(X_train,Y_train)

alg2 = LinearRegression()
alg2.fit(X1_train,Y1_train)
```

```python
Y_pred = alg1.predict(X_test)
train_score1 = alg1.score(X_train,Y_train)
test_score1 = alg1.score(X_test,Y_test)
print(train_score1,test_score1)

Y1_pred = alg2.predict(X1_test)
train_score2 = alg2.score(X1_train,Y1_train)
test_score2 = alg2.score(X1_test,Y1_test)
print(train_score2,test_score2)
```

clearly by taking square of features as another feature our score/accuracy has been improved. Now let's see the effect on score 
by taking product of different features as an extra feature

```python
df2 = df1.copy()
df2.shape
```

```python
k = 0
for i in boston.feature_names:
    for j in range(k,13):
        df2[i+boston.feature_names[j]] = (df2[i])*(df2[boston.feature_names[j]])
    k+=1
df2.shape
```

```python
df2
```

```python
X2 = df2.values
X2.shape
```

```python
from sklearn import model_selection
```

```python
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(X,Y,random_state = 0)
X2_train,X2_test,Y2_train,Y2_test = model_selection.train_test_split(X2,Y,random_state = 0)
X.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
alg1 = LinearRegression()
alg1.fit(X_train,Y_train)

alg3 = LinearRegression()
alg3.fit(X2_train,Y2_train)
```

```python
Y_pred = alg1.predict(X_test)
train_score1 = alg1.score(X_train,Y_train)
test_score1 = alg1.score(X_test,Y_test)
print(train_score1,test_score1)

Y2_pred = alg3.predict(X2_test)
train_score3 = alg3.score(X2_train,Y2_train)
test_score3 = alg3.score(X2_test,Y2_test)
print(train_score3,test_score3)
```

we can see that test_score2 > test_score1 > test_score3. It is because we have balanced our df data by adding some columns and so the efficiency and test score increse but test_score3 is less than test_score1 because we have overfitted our data (so many columns in df2 were added (overfitting)) due to which the algorithm remember our data and its efficiency for training data is superb but it is even worse than normal case for test data.
